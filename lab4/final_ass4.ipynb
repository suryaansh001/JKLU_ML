calcualting fn for cost function 

J(w,b)=1/2m(summation [wx+b - y]**2)

```python
#calculate the cost function
def costfn(x,y,weights):
    m=len(y)
    #multiplying the weights with the x matrix
    cost=(1/(2*m))*np.sum((np.dot(x,weights)-y)**2)
    return cost


# print("x ka shape :",x.shape)
# print("y ka shape :",y.shape)
# print("weights ka shape :",weights.shape)
# print("shape of weights * x",np.dot(x,weights).shape)
# print(costfn(x,y,weights))
```

Based on the cost function now updating the values of weights 

w = w - learning_Rate * (d(J)/d(w))


dj/dw = 1/m * (wx+b -y )*x


```python
def updateweights(x,y,weights,learning_rate):
    
    m=len(y)
    weights=weights-(learning_rate/m)*(np.dot(x.T,(np.dot(x,weights)-y)))
    
    # print("shape of x.T",x.T.shape)
    # print("shape of np.dot(x,weights)-y",(np.dot(x,weights)-y).shape)
    # print(weights)
    return weights
print(updateweights(x,y,weights,learning_rate).shape)
```

now we need to do this untill convergence there comes our gradient descent function 

here convergence is eta =0.0001


```python
def gradient_descent(x,y,weights,learning_rate,total_iter):
    #cost history has to be maintainded to see the cost function
    cost_his=[]
    weights_his=[]
    i=0
    
    # while costfn(x,y,weights)>0.1:
    for i in range(total_iter):
        weights=updateweights(x,y,weights,learning_rate)
        weights_his.append(weights)
        cost_his.append(costfn(x,y,weights))
        # i+=1
        #print("weights after iteration",i,weights)
        
    return weights,cost_his,weights_his

final_weights,cost_his,weights_his=gradient_descent(x,y,weights,learning_rate,total_iter)
print(final_weights)
print(cost_his)
print(weights_his)
```

plotting the garph of cost fn vs no of iterations 


```python
def graph_cost_vs_iter(cost_his,total_iter):
    iter=[i for i in range(total_iter)]
    plt.plot(iter,cost_his)
    plt.xlabel('iterations')
    plt.ylabel('cost')
    plt.title('cost vs iterations')
    plt.show()

graph_cost_vs_iter(cost_his,total_iter)
```
